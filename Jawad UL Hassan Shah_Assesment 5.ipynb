### Part 1: Working with Variables, Input, and Strings

```python
full_name = input("Enter your full name: ")
```

```python
full_name = full_name.strip()
```

```python
print(f"Hello, {full_name}! Welcome to the Python assignment.")
```

```python
#string method
```

```python
print("Your name in uppercase:", full_name.upper())
print("Your name in lowercase:", full_name.lower())
print("Number of characters in your name:", len(full_name))
print("Name with hyphens instead of spaces:", full_name.replace(" ", "-"))
```

```python
birth_year = input("Enter your birth year: ")
```

```python
birth_year = int(birth_year)
```

```python
current_year = 2026
age = current_year - birth_year
print(f"You are approximately {age} years old.")
```

### Part 2: Lists and Indexing

```python
fruits = input("Enter three of your favorite fruits (comma-separated): ")
```

```python
fruit_list = fruits.split(",")
```

```python
print("Your list of fruits:", fruit_list)
```

```python
#List indexing
```

```python
print("First fruit:", fruit_list[0])
print("Last fruit:", fruit_list[-1])
print("Second fruit:", fruit_list[1])
```

```python
#Slicing
```

```python
print("First two fruits:", fruit_list[0:2])
```

```python
#append
```

```python
print("Adding a new fruit...")
fruit_list.append("strawberry")
print("Updated list:", fruit_list)
```

```python
#Remove
```

```python
print("Removing one fruit...")
fruit_list.pop(1)   
print("Updated list:", fruit_list)
```

```python
#Length of list
```

```python
print("Length of the list is now", len(fruit_list))
```

### Part 3 (Optional Challenge): Combining Concepts

```python
full_name = input("Enter your full name: ").strip()
```

```python
letters_list = list(full_name.replace(" ", ""))
```

```python
print("First 3 letters:", letters_list[:3])
print("Last 3 letters:", letters_list[-3:])
```

```python
full_name = ("".join(letters_list[:3]).upper() + "".join(letters_list[3:]).lower())

print("full name:", full_name)
```

```python
numbers = input("Enter 5 numbers (space-separated): ")

numbers_list = list(map(int, numbers.split()))

total = sum(numbers_list)
average = total / len(numbers_list)

print("Sum of your numbers:", total)
print("Average of your numbers:", average)
```
