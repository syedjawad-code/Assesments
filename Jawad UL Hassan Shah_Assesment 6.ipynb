1. ### Write a Python program that can do the following: - You have $50 - You buy an item that is $15, that has a 3% tax - Print how much money you have left, after purchasing the item.

```python
money = 50
item_price = 15
tax_rate = 0.03
```

```python
tax = item_price * tax_rate
total_cost = item_price + tax
```

```python
money_left = money - total_cost
```

```python
print("Money left after purchase:", money_left)
```

2. ### Ask the user how many days until their birthday
- ### Print an approx. number of weeks until their birthday
- ### 1 week is = to 7 days.

```python
days = int(input("How many days until your birthday? "))
weeks = days / 7
```

```python
print("Approximate weeks until your birthday:", weeks)
```

### Create a variable “Grade” holding an integer between 0 to100. Code to print
### the letter grade of the number grade variable.

```python
grade = int(input("Enter your grade (0-100): "))
```

```python
if 90 <= grade <= 100:
    print("Letter Grade: A")
elif 80 <= grade <= 89:
    print("Letter Grade: B")
elif 70 <= grade <= 79:
    print("Letter Grade: C")
elif 60 <= grade <= 69:
    print("Letter Grade: D")
elif 0 <= grade <= 59:
    print("Letter Grade: F")
else:
    print("Invalid grade entered")
```
