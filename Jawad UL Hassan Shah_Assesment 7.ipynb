```python
#5. Write a program to reverse a string using a for loop.
```

```python
name = input("Enter a name")
reversed_name = ""

for i in range(len(name)-1, -1, -1):
    reversed_name += name[i]
    print("Reversed name:", reversed_name)
```

```python
#6. Count the number of vowels present in a string using a for loop.
```

```python
text = input("Enter a name: ")
count = 0
for letter in text:
    if letter.lower() in "aeiou": 
        count += 1

# Print the result
print( count)
```

```python
#8. Write a program to check whether a number is prime using a for loop.
```

```python
num = int(input("Enter a number"))
factors = 0
```

```python
for i in range(2, num):
    factors += (num % i == 0)
```

```python
print(num, "is prime" * (factors == 0) + "not prime" * (factors > 0))
```

```python
# 9. Given a list:   numbers = [12, 5, 7, 18, 21, 30]   Print only numbers divisible by 3.
```

```python
numbers = [12, 5, 7, 18, 21, 30]

for num in numbers:
    if num % 3 == 0:
        print(num)
```

```python
#10.Break and Continue Print numbers from 1 to 20 but: - Skip multiples of 3. - Stop execution if number becomes greater than 15.
```

```python
for i in range(1, 21):
    if i > 15:
        break         
    if i % 3 == 0:
        continue       
    print(i)
```
