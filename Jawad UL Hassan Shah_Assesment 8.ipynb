```python
# 1. Student Grade Calculator
```

```python
def calculate_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 80:
        return "B"
    elif marks >= 70:
        return "C"
    else:
        return "Fail"
```

```python
your_numbers = int(input("Enter your marks"))
grade = calculate_grade(your_numbers)
```

```python
print("Your grade is:", grade)
```

```python
#3. Factorial Using Function
```

```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
```

```python
num = int(input("Enter a number"))
```

```python
print("Factorial of", num, "is:", factorial(num))
```

```python
def find_max(a, b, c):
    return max(a, b, c)
```
